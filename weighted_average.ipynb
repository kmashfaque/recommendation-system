```python
import pandas as pd
```

```python
credits=pd.read_csv("tmdb_5000_credits.csv")
credits.head()
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
movies.head(2)
```

```python
movies.shape
```

```python
credits_column_renamed=credits.rename(index=str,columns={"movie_id":"id"})
```

```python
movies_df_merge=movies.merge(credits_column_renamed,on="id")
movies_df_merge.head()
```

```python
categorical_feature=[feature for feature in movies_df_merge.columns if movies_df_merge[feature].dtypes=="O"]
categorical_feature
```

```python
import numpy as np
na_values=[feature for feature in movies_df_merge.columns if feature not in categorical_feature and movies_df_merge[feature].isnull().sum()>1]

print(movies_df_merge[na_values].isna().sum())
# for feature in na_values:
#     print(np.round(movies_df_merge[feature].isnull().mean(),4)," % missing values")
```

```python
movies_df_merge.shape
```

```python
movies_cleaned_df = movies_df_merge.drop(columns=['homepage', 'title_x', 'title_y', 'status','production_countries'])
movies_cleaned_df.shape
```

```python
movies_cleaned_df.info()
```

```python
avg=movies_cleaned_df["runtime"].mean()

new_avg=np.round(avg)

movies_cleaned_df["runtime"]=movies_cleaned_df["runtime"].fillna(new_avg)
```

```python
na_values=[feature for feature in movies_cleaned_df.columns if feature not in categorical_feature and movies_cleaned_df[feature].isnull().sum()>1]

print(movies_cleaned_df[na_values].isna().sum())
```
